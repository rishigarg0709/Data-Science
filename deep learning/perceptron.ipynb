```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
```

```python
X,Y = make_blobs(n_samples=500, centers=2,n_features=2,random_state=1)
#X,Y = make_moons(n_samples=1000, shuffle=True, noise=0.2, random_state=1)
print(X.shape,Y.shape)
```

```python
plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap= plt.cm.Accent)
plt.show()
```

```python
def sigmoid(z):
    return (1.0/(1+np.exp(-z)))

def predict(w,X):
    z=np.dot(X,w)
    return sigmoid(z)

def gradient(X,Y,w,lrate):
    Y_=predict(w,X)
    grad=np.dot(X.T,Y_-Y)
    w=w-lrate*grad/(float(X.shape[0]))
    return w

def trainmodel(X,Y,lrate=0.5,epochs=100):
    
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    w=np.zeros(X.shape[1])
    
    for i in range(epochs):
        w=gradient(X,Y,w,lrate)      
    return w

#
```

```python
wt=trainmodel(X,Y)
print(wt)
```

```python
x1 = np.linspace(-12,2,10)
x2 = -(wt[0] + wt[1]*x1)/wt[2]

plt.plot(x1,x2)

plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap= plt.cm.Accent)
plt.show()
```
